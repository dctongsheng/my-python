```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
df6 = pd.DataFrame({
    'name':['joe', 'susan', 'anne', 'black', 'monika','ronaldo','leonarldo','tom','yilianna','bulanni'],
    'age':[19,19,18,20,20,18,19,20,18,19],
    'sex':['man','women','women','man','women','man','man','man','women','women'],
    'address':['上海','北京','上海','北京','北京','上海','北京','上海','北京','上海'],
    'money':[8000,8500,7000,9000,10000,7500,8800,9300,12000,11000]
})
df6
```

```python
#根据地区进行分组   查看平均年龄和工资
#groupby函数对pandas进行分组，返回一个可迭代的groupby对象
#然后对对象进行操作
s=df6.groupby(["address","sex"])
for i in s:
    print(i)
```

```python
s.mean()
```

```python
#apply的用法类似于map函数
def f(x):
    return x.mean()
s=df6.groupby(["name"]).apply(f)
s
```
